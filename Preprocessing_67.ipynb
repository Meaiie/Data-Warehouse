* (10 points) จากข้อมูลที่ให้มา เตรียมข้อมูลให้พร้อมโดยการจัดการกับค่าว่างในข้อมูล ให้อธิบายวิธีที่ท่านเลือกใช้ พร้อมยกเหตุผลประกอบ

การเตรียมข้อมูลทำได้หลายวิธี เนื่องจากยังไม่ทราบว่าจะนำข้อมูลไปใช้ในการวิเคราะห์หรือต้องการนำไปใช้ต่อในส่วนใดบ้าง ในกรณีที่บางคอลัมน์ที่มีค่าว่างหรือค่า outlier มากแต่ไม่ได้ใช้งาน เราจึงไม่จำเป็นต้องดรอปค่าว่างที่มีมีอยู่ในคอลัมน์ที่เราไม่ต้องการที่จะใช้งาน ออกไปด้วย

```python
titanic2 = titanic.copy()
```

```python
titanic2
```

เช่น ในกรณีนี้ในข้อต่อไป ต้องการที่จะ ใช้ one-hot encoding ในการแปลงตัวแปร Embarked ซึ่งยังมีค่าว่างอยู่ จึงเลือกที่ดรอปค่าว่างที่มีอยู่ในคอลัมน์นี้ไปเลย เนื่องจากค่าว่างในคอลัมน์นี้ มีเพียง 2 record

```python
titanic2['embarked'] = titanic2['embarked'].dropna()
titanic2
```

เลือกที่จะแทนค่าว่างในคอลัมน์ age เป็นค่าฐานนิยมของอายุ เนื่องจากส่วนตัวเห็นว่าผู้ที่ท่องเที่ยวทางเรือส่วนใหญ่มักจะอยู่ในกลุ่มอายุที่พอ ๆ กัน

```python
age_mode = titanic2['age'].mode()[0]
titanic2['age'].fillna(age_mode, inplace=True)
```

```python
titanic2
```

* แปลงข้อมูลเชิงคุณภาพให้เป็นตัวเลข \
    i. (10 points) ใช้ one-hot encoding ในการแปลงตัวแปร Embarked \
    ii. (10 points) แปลงตัวแปร Sex โดยให้เพศชายเป็น 0 และเพศหญิงเป็น 1 \
    iii. (10 points) สร้างตัวแปร FamilySize ซึ่งเป็นผลรวมของตัวแปร SibSp (number of siblings/spouses aboard) และ Parch (number of parents/children aboard) \
    iv. (10 points) สร้างตัวแปร IsAlone ซึ่งมีค่าเป็น 1 ถ้า FamilySize เป็น 0, และมีค่าเป็นศูนย์เมื่อเป็นอื่น

```python
# One-hot encoding for Embarked
titanic2 = pd.get_dummies(titanic2, columns=['embarked'], prefix=['embarked'])
```

```python
titanic2
```

```python
# Convert 'sex' column to numerical values
titanic2['sex'] = titanic2['sex'].map({'male': 0, 'female': 1})
titanic2
```

```python
# Create FamilySize column
titanic2['FamilySize'] = titanic2['sibsp'] + titanic2['parch']
titanic2
```

```python
# Create IsAlone column
titanic2['IsAlone'] = np.where(titanic2['FamilySize'] == 0, 1, 0)
titanic2
```

* ทำการ Normalize ข้อมูล \
    i. (10 points) ตัวแปร Fare และ Age ใช้ Min-Max Scaling \
    ii. (10 points) สร้างตัวแปรใหม่ Fare_Per_Person ด้วยการนำตัวแปร Fare มาหารด้วย FamilySize + 1 \
    iii. (10 points) สร้างตัวแปร AgeGroup ที่ระบุกลุ่มอายุของตัวอย่าง โดยแบ่งเป็นกลุ่มดังนี้ [0, 12), [12, 18), [18, 50) และ [50, 100)

```python
import pandas as pd
import numpy as np
pd.cut(np.array([1, 7, 5, 4, 6, 3]), 3)
```

```python
pd.cut(np.array([1, 7, 5, 4, 6, 3]), [0,2,5,8])
```

```python
# iii Create AgeGroup column
titanic2['AgeGroup'] = pd.cut(titanic2['age'], bins=[0, 12, 18, 50, 100], labels=['0-12', '12-18', '18-50', '50-100'])
titanic2
```

```python
from sklearn.preprocessing import MinMaxScaler
# i
# Create a MinMaxScaler object
scaler = MinMaxScaler()

# Fit and transform the 'fare' and 'age' columns
titanic2[['fare', 'age']] = scaler.fit_transform(titanic2[['fare', 'age']])
titanic2
```

```python
# Create Fare_Per_Person column
titanic2['Fare_Per_Person'] = titanic2['fare'] / (titanic2['FamilySize'] + 1)
titanic2
```

* (10 points) นำเสนอการแจกแจงหรือทำความเข้าใจตัวแปรสำคัญในข้อมูล เช่น Age, Fare, and Pclass ด้วยการสร้างฮิสโตแกรมหรือแผนภาพกล่อง

```python
# Histogram for Age
plt.figure(figsize=(8, 6))
plt.hist(titanic2['age'], bins=20)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age')
plt.show()
```

```python
# Histogram for Fare
plt.figure(figsize=(8, 6))
plt.hist(titanic2['fare'], bins=20)
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.title('Histogram of Fare')
plt.show()
```

```python
# Box plot for Pclass
plt.figure(figsize=(8, 6))
plt.boxplot(titanic2['pclass'])
plt.xlabel('Pclass')
plt.ylabel('Value')
plt.title('Boxplot of Pclass')
plt.show()
```

* เมื่อเตรียมข้อมูลพร้อมแล้ว ให้หาความสัมพันธ์ของตัวแปรต่อไปนี้จากข้อมูล ซึ่งอาจนำเสนอด้วยสถิติและแผนภูมิหรือกราฟก็ได้: \
    i. (10 points) survived และ pclass \
    ii. (10 points) survived และ sex \
    iii. (10 points) survived และ age \
    iv. (10 points) จำนวนสมาชิกครอบครัวที่รอดชีวิต และ Fare

```python
# Calculate survival rate by Pclass
survival_rate_by_pclass = titanic2.groupby('pclass')['survived'].mean()
print(survival_rate_by_pclass)

# Create a bar chart
plt.figure(figsize=(8, 6))
plt.bar(survival_rate_by_pclass.index, survival_rate_by_pclass.values)
plt.xlabel('Pclass')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Pclass')
plt.show()
```

จะเห็นว่า Pclass ที่ 1 2 และ 3 จะมีอัตราการรอดชีวิตจากมากไปน้อยตามลำดับ

```python
# Calculate survival rate by sex
survival_rate_by_sex = titanic2.groupby('sex')['survived'].mean()
print(survival_rate_by_sex)

# Create a bar chart
plt.figure(figsize=(8, 6))
plt.bar(survival_rate_by_sex.index, survival_rate_by_sex.values)
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Sex')
plt.show()
```

จากกราฟ แสดงให้เห็นว่าอัตราการรอดชีวิตของเพศหญิงมีค่าที่สูงกว่าเพศชาย

```python
# Calculate survival rate by age
survival_rate_by_age = titanic2.groupby('AgeGroup')['survived'].mean()
print(survival_rate_by_age)

# Create a bar chart
plt.figure(figsize=(8, 6))
plt.bar(survival_rate_by_age.index, survival_rate_by_age.values)
plt.xlabel('Age Group')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Age Group')
plt.show()
```

จากกราฟแสดงให้เห็นว่าอัตราการรอดชีวิตของผู็โดยสารที่อายุอยู่ในช่วง 0 -12 ปีมีโอกาสรอดชีวิตสูงกว่าช่วงอายุอื่น ๆ

* (10 points) สร้างตารางสรุปข้อมูลเพื่อหาคุณลักษณะของผู้โดยสารเรือไททานิค และศึกษาเกี่ยวกับการเสียชีวิตของผู้โดยสาร ทำการสรุปผลที่ได้จากตาราง

```python
pivot_table = pd.pivot_table(titanic2, values='survived', index=['sex', 'pclass'], columns=['AgeGroup'], aggfunc='mean')
pivot_table
```

จะเห็นว่าในเพศชายจะมีอัตราการเสียชีวิตมากกว่าผู้โดยสารเพศหญิง รวมถึงคลาสของผู้โดยสารที่ดีกว่าก็จะมีอัตราการรอดชีวิตที่สูงกว่า ในที่นี้ pclass ที่ 1 คือคลาสที่หรูที่สุด อาจมีสิ่งอำนวยความสะดวกหรือความปลอดภัยที่ดีกว่าผู้โดยสารคลาสอื่น ๆ ตามลำดับ
อัตราการรอดชีวิตในผู้โดยสารอายุ 0-12 ปีจะมีมากกว่าในกลุ่มอายุอืน ที่มีอัตรารอดชีวิตที่ไม่ต่างกันมากนัก
